```python
import pandas as pd
```

```python
# import csv
df = pd.read_csv('/Users/isaacmin/Desktop/lol_code/topplayers.csv')

# print the first 10 lines of the dataframe
df.head(10)
```

```python
def get_routing_value(region):

    # map regions to routing values

    if region in ['na1', 'br1', 'la1', 'la2']:
        return 'AMERICAS'
    elif region in ['kr', 'jp1']:
        return 'ASIA'
    elif region in ['eun1', 'euw1', 'tr1']:
        return 'EUROPE'
    elif region in ['oc1', 'ph2', 'sg2', 'th2', 'tw2', 'vn2']:
        return 'SEA'
    else:
        return 'UNKNOWN'

# apply function
df['routingValue'] = df['region'].apply(get_routing_value)

df.to_csv('updated_topplayers.csv', index = False)
```
