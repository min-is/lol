```python
import pandas as pd
```

```python
df = pd.read_csv(f'/users/isaacmin/Desktop/lol_code/matchids_cleaned.csv')

df.head(10)
```

```python
def shard(region):

    # map regions to routing values

    if region in ['na1', 'br1', 'la1', 'la2']:
        return 'AMERICAS'
    elif region in ['kr', 'jp1']:
        return 'ASIA'
    elif region in ['eun1', 'euw1', 'tr1']:
        return 'EUROPE'
    elif region in ['oc1', 'ph2', 'sg2', 'th2', 'tw2', 'vn2']:
        return 'SEA'
    else:
        return 'UNKNOWN'

# apply function
df['shard'] = df['region'].apply(shard)

df.to_csv('match_ids_cl_shard.csv', index = False)
```
