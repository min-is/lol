```python
def get_match_data(matchid, shard, api_key):

    # Endpoint for match data 
    request_url = f'https://{shard}.api.riotgames.com/lol/match/v5/matches/{matchid}'

    # HTTP header
    headers = {
        "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:96.0) Gecko/20100101 Firefox/96.0",
        "Accept-Language": "en-US,en;q=0.9",
        "Accept-Charset": "utf-8",
        "Origin": "https://developer.riotgames.com",
        "X-Riot-Token": api_key
    }

    # rate limits
    time.sleep(1.2)

    response = requests.get(request_url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f'Error fetching data for match ID {matchid}: {response.text}')
        return None

# Load the CSV file
print("Loading CSV file...")
data = pd.read_csv('/Users/isaacmin/Desktop/lol_code/match_ids_cl_shard.csv')

# Initialize an empty list to store match data dictionaries
match_data_list = []

import time

# Initialize counters
request_count = 0
start_time = time.time()

print("Fetching match data...")
for index, row in data.iterrows():
    match_ids = eval(row['match_id'])  # Convert the string representation of list to actual list
    for match_id in match_ids:
        print(f"Fetching data for match ID: {match_id}, Shard: {row['shard']}")
        
        # Check rate limit and adjust sleep accordingly
        request_count += 1
        if request_count % 20 == 0 and time.time() - start_time < 1:
            time.sleep(1 - (time.time() - start_time))  # Wait to maintain rate limit
        
        if request_count % 100 == 0 and time.time() - start_time < 120:
            time.sleep(120 - (time.time() - start_time))  # Wait for 2 minutes to maintain rate limit
            start_time = time.time()  # Reset start time after waiting
        
        match_data = get_match_data(match_id, row['shard'], api_key)
        if match_data:
            match_data_list.append(match_data)
        else:
            print(f'Failed to fetch data for match ID {match_id}')


# Create a DataFrame from the list of match data dictionaries
print("Creating DataFrame...")
df = pd.DataFrame(match_data_list)

# Export the DataFrame to a CSV file
print("Exporting DataFrame to CSV file...")
df.to_csv('/Users/isaacmin/Desktop/lol_code/match_data.csv', index=False)

print("CSV file exported successfully.")
```
